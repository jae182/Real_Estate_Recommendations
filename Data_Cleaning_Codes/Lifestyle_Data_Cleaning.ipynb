```python
# Import necessary libraries
import pandas as pd
from pathlib import Path
```

```python
# Create Path to Lifestyle data
lifestyle_path = Path('Raw_Data/Lifestyle_Data_by_City.csv')
```

```python
# Read in Lifestyle data
lifestyle_data = pd.read_csv(lifestyle_path)
```

```python
# Check lifestyle_data
lifestyle_data.head()
```
