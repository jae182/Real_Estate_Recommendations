```python
# Import necessary
import pandas as pd
from pathlib import Path
```

```python
# Set Path to data
housing_path = Path('Raw_Data/Housing_Data.csv')
rent_path = Path('Raw_Data/Fair_Market_Rent_Data.csv')
```

```python
# Read in CSV files
housing_data = pd.read_csv(housing_path)
rent_data = pd.read_csv(rent_path)
```

```python
# Check housing_data
housing_data.head()
```

```python
# Rename columns names for rent_data
rent_columns = ['id', 'studio rent', '1 bdrm rent', '2 bdrm rent', '3 bdrm rent', '4 bdrm rent', 'state id', 'metro id', 'area name', 'county code', 'cousub', 'county name', 'county town name', 'population', 'state', 'metro']
rent_data.columns = rent_columns
rent_data.head()
```

```python
# Drop unneccessary columns from housing_data
housing_data = housing_data.drop(columns=['Unadjusted Median Value', 'Years Worth of Sales Data*', 'Number of Sales for Unadjusted Median**', 'Years Worth of Sales Data*.1', 'Number of Sales for Unadjusted Median**.1', 'Geographic Area Used.1', 'Metropolitan/FMR Area Name.1'])
housing_data.head()
```

```python
# Drop unneccessary columns from housing_data
housing_data = housing_data.drop(columns=['Geographic Area Used', 'Unadjusted Median Value.1', 'state2', 'fipscounty', 'metro_code'])
housing_data.head()
```

```python
# Rename Unnamed:20 column
housing_data = housing_data.rename(columns={'Unnamed: 20': 'metro id'})
housing_data.head()
```

```python
# Drop unneccessary columns for rent_data
rent_data = rent_data.drop(columns=['id', 'state id', 'county code', 'cousub', 'county town name', 'population', 'metro'])
rent_data.head()
```

```python
# Check for null values in rent_data
rent_data.isnull().sum()
```

```python
# Check for null values in housing_data
housing_data.isnull().sum()
```

```python
# Set metro code as index for both tables
housing_data = housing_data.set_index('metro id')
rent_data = rent_data.set_index('metro id')
```

```python
# Join housing_data and rent_data tables
housing_cost_df = pd.merge(housing_data, rent_data, on='metro id', how='inner')
housing_cost_df.head()
```

```python
# Drop repeat columns
housing_cost_df = housing_cost_df.drop(columns=['area name', 'county name', 'state'])
housing_cost_df.tail()
```

```python
# Create a list of the city names to search for in the Metropolitan/FMR Area Name column of the housing_cost_df
search_for = ['New York',
              'Los Angeles',
              'Chicago',
              'Houston',
              'Phoenix',
              'Philadelphia',
              'San Antonio',
              'San Diego',
              'Dallas',
              'San Jose',
              'Austin',
              'Jacksonville',
              'Fort Worth',
              'Columbus',
              'Indianapolis',
              'Charlotte',
              'San Francisco',
              'Seattle',
              'Denver',
              'Washington D.C.',
              'Nashville',
              'Oklahoma City',
              'El Paso',
              'Boston',
              'Portland'
]

# Search for names of cities in housing_cost_df 
filtered_housing_df = housing_cost_df[(housing_cost_df['Metropolitan/FMR Area Name'].str.contains('|'.join(search_for)))]

# Check new DataFrame
filtered_housing_df
```

```python
# Check for duplicate rows
filtered_housing_df.duplicated()
```

```python
city_housing_df = filtered_housing_df.drop_duplicates().copy()
city_housing_df
```

```python
# Create list for state abbreviations for 25 most populous cities
st_search_for = ['NY',
                 'CA',
                 'TX',
                 'AZ',
                 'PA',
                 'FL',
                 'OH',
                 'IN',
                 'IL',
                 'NC',
                 'WA',
                 'CO',
                 'TN',
                 'OK',
                 'MA',
                 'OR',
                 'DC'
]

# Filter data for states that contain the appropriate cities (This is because some city names are located in multiple states)
clean_housing_df = city_housing_df[(city_housing_df['State'].str.contains('|'.join(st_search_for)))]
clean_housing_df
```

```python
# For simplicity, drop 1-Unit.1, 2-unit.1, 3-unit.1 4-unit.1
clean_housing_df = clean_housing_df.drop(columns=['1-Unit.1', '2-unit.1', '3-unit.1', '4-unit.1'])
clean_housing_df
```

```python
# There are still rows not pertaining to appropriate cities due to naming similarities. Dataset is much smaller and can filter out unneccessary cities by index
clean_housing_data = clean_housing_df.reset_index()
clean_housing_data = clean_housing_data.set_index('Metropolitan/FMR Area Name')
clean_housing_data = clean_housing_data.drop(['Columbus, IN MSA', 
                                              'Jacksonville, NC MSA', 
                                              'Columbus County, NC', 
                                              'Houston County, TN', 
                                              'Austin County, TX HUD Metro FMR Area', 
                                              'Houston County, TX'])
clean_housing_data
```

```python
# Save table as CSV file in Clean_Data folder
clean_housing_data.to_csv('Clean_Data/clean_housing_data.csv', index=False)
```
