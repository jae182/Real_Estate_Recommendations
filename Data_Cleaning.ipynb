```python
# Import necessary
import pandas as pd
from pathlib import Path
```

```python
# Set Path to data
housing_path = Path('Raw_Data/Housing_Data.csv')
rent_path = Path('Raw_Data/Fair_Market_Rent_Data.csv')
```

```python
# Read in CSV files
housing_data = pd.read_csv(housing_path)
rent_data = pd.read_csv(rent_path)
```

```python
# Check housing_data
housing_data.head()
```

```python
# Rename columns names for rent_data
rent_columns = ['id', 'studio rent', '1 bdrm rent', '2 bdrm rent', '3 bdrm rent', '4 bdrm rent', 'state id', 'metro id', 'area name', 'county code', 'cousub', 'county name', 'county town name', 'population', 'state', 'metro']
rent_data.columns = rent_columns
rent_data.head()
```

```python
# Drop unneccessary columns from housing_data
housing_data = housing_data.drop(columns=['Unadjusted Median Value', 'Years Worth of Sales Data*', 'Number of Sales for Unadjusted Median**', 'Years Worth of Sales Data*.1', 'Number of Sales for Unadjusted Median**.1', 'Geographic Area Used.1', 'Metropolitan/FMR Area Name.1'])
housing_data.head()
```

```python
# Drop unneccessary columns from housing_data
housing_data = housing_data.drop(columns=['Geographic Area Used', 'Unadjusted Median Value.1', 'state2', 'fipscounty', 'metro_code'])
housing_data.head()
```

```python
# Rename Unnamed:20 column
housing_data = housing_data.rename(columns={'Unnamed: 20': 'metro id'})
housing_data.head()
```

```python
# Drop unneccessary columns for rent_data
rent_data = rent_data.drop(columns=['id', 'state id', 'county code', 'cousub', 'county town name', 'population', 'metro'])
rent_data.head()
```

```python
# Check for null values in rent_data
rent_data.isnull().sum()
```

```python
# Check for null values in housing_data
housing_data.isnull().sum()
```

```python
# Set metro code as index for both tables
housing_data = housing_data.set_index('metro id')
rent_data = rent_data.set_index('metro id')
```

```python
# Join housing_data and rent_data tables
housing_cost_df = pd.merge(housing_data, rent_data, on='metro id', how='inner')
housing_cost_df.head()
```

```python
# Drop repeat columns
housing_cost_df = housing_cost_df.drop(columns=['area name', 'county name', 'state'])
housing_cost_df.tail()
```

```python
filtered_housing_df = housing_cost_df[(housing_cost_df['Metropolitan/FMR Area Name'].str.contains('New York'))
                                      |(housing_cost_df.isin(['Los Angeles', 'Chicago']))
                                     ]
filtered_housing_df
```
